```python
def binary_step(x):
    if x<0:
        return 0
    else:
        return 1
binary_step(5),binary_step(-1)
```

```python
def lin_func(x):
    return 4*x
lin_func(4),lin_func(-2)
```

```python
import numpy as np

def sig(x):
    return 1/(1+np.exp(-x))
x=1.0
print(x,sig(x))
x=-10.0
print(x,sig(x))
x=0.0
print(x,sig(x))
x=15.0
print(x,sig(x))
x=-2.0
print(x,sig(x))
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-10,10,50)
p=sig(x)
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.plot(x,p)
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)

x=1.0
print(x, tanh(x))
x=-10.0
print(x, tanh(x))
x=0.0
print(x, tanh(x))
x=15.0
print(x,tanh(x))
x=-2.0
print(x,tanh(x))
```

```python
def relu(x):
    return max(0.0,x)

x=1.0
print(x,relu(x))
x=-10.0
print(x,relu(x))
x=0.0
print(x,relu(x))
x=15.0
print(x,relu(x))
x=-2.0
print(x,relu(x))
```

```python
def leaky_relu(x):
    if x>0:
        return x
    else:
        return 0.01*x

x=1.0
print(x,leaky_relu(x))
x=-10.0
print(x,leaky_relu(x))
x=0.0
print(x,leaky_relu(x))
x=15.0
print(x,leaky_relu(x))
x=-2.0
print(x,leaky_relu(x))
```

```python
def softmax(x):
    exp_scores=np.exp(x)
    return exp_scores/np.sum(exp_scores)

x=1.0
print(x,softmax(x))
x=-10.0
print(x,softmax(x))
x=0.0
print(x,softmax(x))
x=15.0
print(x,softmax(x))
x=-2.0
print(x,softmax(x))
```
