```python
import numpy as np
import matplotlib.pyplot as plt

class Neuron:
    def __init__(self, num_inputs):
        self.weights_hebb = np.random.rand(num_inputs)
        self.weights_cov = np.random.rand(num_inputs)

    def activate(self, inputs, weights):
        return np.dot(inputs, weights)

    def learn_hebbian(self, inputs, lr):
        y = self.activate(inputs, self.weights_hebb)
        self.weights_hebb += lr * y * inputs

    def learn_covariance(self, inputs, lr):
        y = self.activate(inputs, self.weights_cov)
        y_mean = np.mean(y)
        x_mean = np.mean(inputs)
        self.weights_cov += lr * (y - y_mean) * (inputs - x_mean)

# Example usage
num_inputs = 3
inputs = np.array([0.5, 0.3, 0.2])
lr = 0.01
iterations = 200

neuron = Neuron(num_inputs)

hebbian_history = []
cov_history = []

for _ in range(iterations):
    neuron.learn_hebbian(inputs, lr)
    neuron.learn_covariance(inputs, lr)
    hebbian_history.append(np.linalg.norm(neuron.weights_hebb))
    cov_history.append(np.linalg.norm(neuron.weights_cov))

# Plot comparison
plt.plot(hebbian_history, label="Hebbian Learning")
plt.plot(cov_history, label="Covariance Learning")
plt.xlabel("Iterations")
plt.ylabel("Weight Magnitude")
plt.legend()
plt.title("Hebbian vs Covariance Learning")
plt.show()
```

```python
import numpy as np

class Neuron:
    def __init__(self, num_inputs):
        # Initialize weights randomly
        self.weights = np.random.rand(num_inputs)

    def activate(self, inputs):
        # Compute the weighted sum of inputs
        weighted_sum = np.dot(inputs, self.weights)
        return weighted_sum

    def learn_hebbian(self, inputs, learning_rate):
        # Compute the activation
        activation = self.activate(inputs)
        # Update weights using Hebbian learning rule
        self.weights += learning_rate * activation * inputs


# Example usage
if __name__ == "__main__":
    # Create a neuron with 3 inputs
    num_inputs = 3
    neuron = Neuron(num_inputs)

    # Inputs for the neuron
    inputs = np.array([0.5, 0.3, 0.2])

    # Learning rate
    learning_rate = 0.1

    # Number of iterations for learning
    num_iterations = 1000

    # Learn using Hebbian learning algorithm
    for i in range(num_iterations):
        neuron.learn_hebbian(inputs, learning_rate)

    # Print the learned weights
    print("Learned weights:", neuron.weights)
```
